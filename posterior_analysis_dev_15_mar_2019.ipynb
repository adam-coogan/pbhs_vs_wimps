```python
import matplotlib.pyplot as plt
import numpy as np
from posterior_inference_point import PointSourcePosterior, Prior_sv
from posterior_inference import load_p_gamma
import os
```

```python
post_ps = PointSourcePosterior(0.5, 82)
```

```python
svs = np.geomspace(1e-43, 1e-23, 50)
m_dms = np.array([1e1, 1e2, 1e3, 1e4])  # np.geomspace(10, 500, 3)
```

```python
post_ps.save_posterior_table(svs, m_dms, "trapz")
```

```python
post_ps.save_normalized_posterior_table(method="trapz")
```

```python
svs, m_dms, un_post_vals = post_ps.load_posterior(normalized=False)

# Posterior
m_dm_mg, sv_mg = np.meshgrid(m_dms, svs)
post_pcmesh = plt.pcolormesh(m_dm_mg, sv_mg, un_post_vals,
                             edgecolor="face")
plt.colorbar(post_pcmesh)

# Bounds
m_dms, sv_bounds = post_ps.sv_bounds(save=False)
plt.plot(m_dms, sv_bounds, 'r')

plt.xscale("log")
plt.yscale("log")
plt.ylim(1e-45, 1e-25)
```
